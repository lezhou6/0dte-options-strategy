```python
import pandas as pd
from thetadata import ThetaClient
import os
import glob
from datetime import date
```

```python
# get SPY data from 2026-05-06 to 2026-05-19
client = ThetaClient(dataframe_type='pandas')
symbol = 'SPY'
trading_days = pd.bdate_range(end='2026-05-19', periods=10).date
for exp in trading_days:
    out_path = f"../data/raw/greeks/{symbol}/{exp}.parquet"
    if os.path.exists(out_path):
        continue
    
    df = client.option_history_greeks_first_order(
        symbol=symbol,
        expiration=exp,
        start_date=exp,
        end_date=exp,
        interval='15m',
    )
    
    os.makedirs(os.path.dirname(out_path), exist_ok=True)
    df.to_parquet(out_path, index=False)
    print(f"Saved {exp}: {len(df)} rows")
```
